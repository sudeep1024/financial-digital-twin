Load Beta (from CSV):

```python
import pandas as pd

# Load beta estimated earlier
beta_df = pd.read_csv("../data/beta_hdfc.csv")
beta_df
```

Define WACC Inputs:

```python
# Core financial inputs
risk_free_rate = 0.07          # 7% (India 10Y approx)
market_risk_premium = 0.06     # 6%
cost_of_debt = 0.08            # 8%
tax_rate = 0.25                # 25%

# Capital structure (can later be dynamic)
debt_weight = 0.30
equity_weight = 0.70

beta = beta_df["beta"].iloc[0]
```

Cost of Equity (CAPM):

Cost of Equity = Rf​+β(Rm​−Rf​)

```python
cost_of_equity = risk_free_rate + beta * market_risk_premium
cost_of_equity

```

Compute WACC:

WACC = E/V⋅Re ​+ D/V⋅Rd​⋅(1−T)

```python
wacc = (
    equity_weight * cost_of_equity +
    debt_weight * cost_of_debt * (1 - tax_rate)
)

wacc
```

Make WACC Dynamic:

```python
wacc_df = pd.DataFrame({
    "ticker": ["HDFCBANK.NS"],
    "beta": [beta],
    "risk_free_rate": [risk_free_rate],
    "market_premium": [market_risk_premium],
    "cost_of_debt": [cost_of_debt],
    "tax_rate": [tax_rate],
    "equity_weight": [equity_weight],
    "debt_weight": [debt_weight],
    "wacc": [wacc]
})

wacc_df
```

Save WACC:

```python
wacc_df.to_csv("../data/wacc_hdfc.csv", index=False)
```
