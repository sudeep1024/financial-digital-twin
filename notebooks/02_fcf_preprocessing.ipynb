Import Libraries:

```python
import pandas as pd
import numpy as np
import os
```

Load Financial Statements:

```python
BASE_DIR = "../data/fundamentals"

income = pd.read_csv(f"{BASE_DIR}/income_statement.csv", index_col=0)
balance = pd.read_csv(f"{BASE_DIR}/balance_sheet.csv", index_col=0)
cashflow = pd.read_csv(f"{BASE_DIR}/cashflow_statement.csv", index_col=0)

income.head(), cashflow.head()
```

Inspect Cash Flow Columns:

```python
cashflow.columns
```

Compute Free Cash Flow (FCF):

```python
cashflow["FCF"] = (
    cashflow["Operating Cash Flow"] -
    cashflow["Capital Expenditure"]
)
```

```python
cashflow[["Operating Cash Flow", "Capital Expenditure", "FCF"]]
```

Handle Missing Values:

```python
cashflow = cashflow.dropna(subset=["FCF"])
```

Sort by Time:

```python
cashflow = cashflow.sort_index()
```

Save Clean FCF Data:

```python
fcf_df = cashflow[["FCF"]]
fcf_df.to_csv("../data/fundamentals/fcf_clean.csv")

fcf_df
```
