Imports:

```python
import pandas as pd
import numpy as np
import yfinance as yf
import os
from datetime import datetime

```

Create Data Folders:

```python
BASE_DIR = "../data"
PRICE_DIR = os.path.join(BASE_DIR, "prices")
FUND_DIR = os.path.join(BASE_DIR, "fundamentals")

os.makedirs(PRICE_DIR, exist_ok=True)
os.makedirs(FUND_DIR, exist_ok=True)
```

Download HDFC Bank Prices:

```python
TICKER = "HDFCBANK.NS"
START_DATE = "2000-01-01"
END_DATE = datetime.today().strftime("%Y-%m-%d")

stock = yf.download(TICKER, start=START_DATE, end=END_DATE)
stock.to_csv(f"{PRICE_DIR}/HDFCBANK_prices.csv")

stock.head()
```

Download NIFTY Prices:

```python
INDEX = "^NSEI"

index_data = yf.download(INDEX, start=START_DATE, end=END_DATE)
index_data.to_csv(f"{PRICE_DIR}/NIFTY_prices.csv")

index_data.head()
```

Download Financial Statements:

```python
company = yf.Ticker(TICKER)

income_stmt = company.financials.T
balance_sheet = company.balance_sheet.T
cashflow_stmt = company.cashflow.T
```

Save Statements:

```python
income_stmt.to_csv(f"{FUND_DIR}/income_statement.csv")
balance_sheet.to_csv(f"{FUND_DIR}/balance_sheet.csv")
cashflow_stmt.to_csv(f"{FUND_DIR}/cashflow_statement.csv")
```
