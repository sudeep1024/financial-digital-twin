```python
import pandas as pd
import numpy as np
```

Load Monte Carlo results:

```python
mc_df = pd.read_csv("../data/monte_carlo_valuation.csv")
simulated_values = mc_df["valuation"].values
```

Load base DCF:

```python
dcf_df = pd.read_csv("../data/dcf_hdfc.csv")
base_dcf = dcf_df.loc[0, "enterprise_value"]
```

Compute risk statistics:

```python
p10, p50, p90 = np.percentile(simulated_values, [10, 50, 90])

prob_undervalued = np.mean(simulated_values > base_dcf)
prob_overvalued = np.mean(simulated_values < base_dcf)

upside_pct = (p90 / base_dcf - 1) * 100
downside_pct = (p10 / base_dcf - 1) * 100
```

Create final valuation summary table:

```python
risk_summary_df = pd.DataFrame({
    "metric": [
        "Base DCF Value",
        "P10 (Bear Case)",
        "P50 (Median)",
        "P90 (Bull Case)",
        "Upside to P90 (%)",
        "Downside to P10 (%)",
        "Probability Undervalued",
        "Probability Overvalued"
    ],
    "value": [
        base_dcf,
        p10,
        p50,
        p90,
        upside_pct,
        downside_pct,
        prob_undervalued,
        prob_overvalued
    ]
})

risk_summary_df
```

save the data for later uses:

```python
risk_summary_df.to_csv("../data/risk_summary_hdfc.csv", index=False)
```
